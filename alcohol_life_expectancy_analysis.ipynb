```python
# –≥—Ä—É–ø–ø–∏—Ä—É–µ–º –ø–æ —Å—Ç—Ä–∞–Ω–∞–º –∏ —Ç–∏–ø—É –Ω–∞—É—á–Ω–æ–π —Å—Ç–∞—Ç—å–∏ –∏ –≤—ã—á–∏—Å–ª—è–µ–º –∞–≥–≥—Ä–µ–≥–∏—Ä—É—é—â–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏

grouped = merged_df.groupby(['CountryCode', 'GhoCode']).agg(
    life_mean=('Numeric', 'mean'),
    life_median=('Numeric', 'median'),
    life_min=('Numeric', 'min'),
    life_max=('Numeric', 'max'),
    alcohol_mean=('total_litres_of_pure_alcohol', 'mean')
).reset_index()
grouped
```

–î–ª—è –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∂–∏–∑–Ω–∏ (Numeric) –±–µ—Ä—É mean, median, min –∏ max, —á—Ç–æ–±—ã –≤–∏–¥–µ—Ç—å —Å—Ä–µ–¥–Ω—é—é –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å, —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏ —Ä–∞–∑–±—Ä–æ—Å.

–î–ª—è –ø–æ—Ç—Ä–µ–±–ª–µ–Ω–∏—è –∞–ª–∫–æ–≥–æ–ª—è (total_litres_of_pure_alcohol) –±–µ—Ä—É mean, —Ç–∞–∫ –∫–∞–∫ –∏–Ω—Ç–µ—Ä–µ—Å–Ω–æ —Å—Ä–µ–¥–Ω–µ–µ –ø–æ—Ç—Ä–µ–±–ª–µ–Ω–∏–µ –Ω–∞ —á–µ–ª–æ–≤–µ–∫–∞ –≤ –∫–∞–∂–¥–æ–π —Å—Ç—Ä–∞–Ω–µ –∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ç–∏–ø–∞ —Å—Ç–∞—Ç—å–∏.

*–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ù–µ –∑–∞–±—É–¥—å—Ç–µ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –≤—Å–µ –∏–∑–º–µ–Ω–µ–Ω–∏—è! –í –ø–æ—Å–ª–µ–¥—É—é—â–∏—Ö –∑–∞–¥–∞–Ω–∏—è—Ö –≤—ã –¥–æ–ª–∂–Ω—ã —Ä–∞–±–æ—Ç–∞—Ç—å —Å –æ–±–Ω–æ–≤–ª–µ–Ω–Ω—ã–º –¥–∞—Ç–∞—Å–µ—Ç–æ–º.*

```python
# —Ä–µ—à–∏–ª–∏ –≤—Å—Ç–∞–≤–∏—Ç—å life_mean –∏ alcohol_mean
columns_to_merge = grouped[['CountryCode', 'GhoCode', 'life_mean', 'alcohol_mean']]
life_exp = life_exp.merge(
    columns_to_merge,
    on=['CountryCode', 'GhoCode'],
    how='left'
)
life_exp = life_exp.rename(columns={
    'life_mean': 'avg_life_by_country',
    'alcohol_mean': 'avg_alcohol_by_country'
})
life_exp
```

```python
# –°–Ω–∞—á–∞–ª–∞ —Å–æ–∑–¥–∞—ë–º –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —Å—Ä–µ–¥–Ω–µ–µ –ø–æ—Ç—Ä–µ–±–ª–µ–Ω–∏–µ –∞–ª–∫–æ–≥–æ–ª—è –Ω–∞ —á–µ–ª–æ–≤–µ–∫–∞ –ø–æ –≤—Å–µ–º –≤–∏–¥–∞–º
alcohol_cols = ['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']
merged_df['total_servings'] = merged_df[alcohol_cols].mean(axis=1)
# –°—á–∏—Ç–∞–µ–º –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –º–µ–∂–¥—É –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å—é –∂–∏–∑–Ω–∏ (Numeric) –∏ –≤—Å–µ–º–∏ –≤–∏–¥–∞–º–∏ –ø–æ—Ç—Ä–µ–±–ª–µ–Ω–∏—è –∞–ª–∫–æ–≥–æ–ª—è
corrs = merged_df[['Numeric'] + alcohol_cols].corr()['Numeric'].sort_values(ascending=False)
corrs
```

```python
# –°—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω—é—é –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∂–∏–∑–Ω–∏ –∏ —Å—Ä–µ–¥–Ω–µ–µ –ø–æ—Ç—Ä–µ–±–ª–µ–Ω–∏–µ –∞–ª–∫–æ–≥–æ–ª—è –ø–æ —Å—Ç—Ä–∞–Ω–∞–º –∏ –ø–æ–ª—É
grouped = merged_df.groupby(['CountryCode','SexCode']).agg({
    'Numeric':'mean',
    'total_litres_of_pure_alcohol':'mean'
}).reset_index()

# –û–ø—Ä–µ–¥–µ–ª—è–µ–º —Ä–∞–∑–Ω–∏—Ü—É –≤ –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∂–∏–∑–Ω–∏ –º–µ–∂–¥—É –∂–µ–Ω—â–∏–Ω–∞–º–∏ –∏ –º—É–∂—á–∏–Ω–∞–º–∏ –≤ –∫–∞–∂–¥–æ–π —Å—Ç—Ä–∞–Ω–µ
grouped['female_male_diff'] = grouped.groupby('CountryCode')['Numeric'].transform(lambda x: x.max() - x.min())

# –°–º–æ—Ç—Ä–∏–º 10 —Å—Ç—Ä–∞–Ω —Å —Å–∞–º–æ–π –±–æ–ª—å—à–æ–π —Ä–∞–∑–Ω–∏—Ü–µ–π
grouped.sort_values('female_male_diff', ascending=False).head(10)
```
